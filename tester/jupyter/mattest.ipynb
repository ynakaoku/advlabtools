```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
#import pandas as pd
import os
import commands
import json

f = open('test01-iperf3-180314-194540/test01-cl-172.16.130.201.json', 'r')
perf_dict = json.load(f)

plt.rcParams["font.size"] = 22
plt.rcParams["xtick.labelsize"] = 12
plt.rcParams["ytick.labelsize"] = 15
plt.rcParams["legend.fontsize"] = 15
plt.grid()

plt.xlabel("Interval")
plt.ylabel("Throughput()")

x = np.array(range(len(perf_dict["intervals"])))
y = np.array([])

for p in perf_dict["intervals"]:
  y = np.append(y, p["sum"]["bits_per_second"])

plt.plot(x, y, "r",marker="o",markersize=2)
```
